```python
import pandas as pd

# Load dataset
df = pd.read_csv("data/parkinsons.data")

# Drop the 'name' column
df.drop(columns=["name"], inplace=True)

# Show first few rows
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df.drop("status", axis=1)
y = df["status"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42)

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score

model = SVC(kernel="rbf", probability=True, random_state=42)
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)
y_proba = model.predict_proba(X_test_scaled)[:, 1]

print("Accuracy:", accuracy_score(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_proba))
print(classification_report(y_test, y_pred))
```

```python
from joblib import dump

dump((model, scaler), "parkinsons_model.joblib")
```

```python
from joblib import load
import numpy as np

# Load model and scaler
model, scaler = load("parkinsons_model.joblib")

# Example input (replace values or take from df.sample(1))
sample = X.iloc[0].values.reshape(1, -1)
sample_scaled = scaler.transform(sample)

# Predict
pred = model.predict(sample_scaled)[0]
prob = model.predict_proba(sample_scaled)[0, 1]

print("Prediction:", "Parkinson’s" if pred == 1 else "Healthy")
print("Confidence:", round(prob * 100, 2), "%")
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay, RocCurveDisplay

ConfusionMatrixDisplay.from_estimator(model, X_test_scaled, y_test)
plt.title("Confusion Matrix")
plt.show()

RocCurveDisplay.from_estimator(model, X_test_scaled, y_test)
plt.title("ROC Curve")
plt.show()
```
